<h1>[ ML_04_Decision Tree ]</h1>

<h2># Import Library</h2>

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

<h2># Train Data : Read & Split</h2>

```python
train = pd.read_csv("./iris-3.csv")
print(train)
trainData = np.array(train)

# split data, target
x_train = trainData[:, 0:4]
y_train = trainData[:, 4]
```

<h2># Fitting Decision Tree</h2>

```python
# 클래스 객체 생성
tree = DecisionTreeClassifier(max_depth = 3, random_state = 0)

# 모델 생성
# x_train : 훈련데이터 numpy 배열
# y_train : 훈련데이터 레이블
tree.fit(x_train, y_train)
```

<h2># Graphviz : Drawing Graph</h2>

```python
# graphviz
import graphviz
from sklearn.tree import export_graphviz

featureName = ['feature1', 'feature2', 'feature3', 'feature4']
targetName = ['1', '2', '3']
export_graphviz(tree, out_file="tree_hw.dot", class_names=targetName)

with open("tree_hw.dot") as f:
    dot_graph = f.read()
display(graphviz.Source(dot_graph))
```
